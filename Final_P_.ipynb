```python
import numpy as np
from scipy.integrate import simps
def simpsons_rule(x, y):
    n = len(x) - 1
    if n % 2 == 1:
        raise ValueError("Кількість інтервалів повинна бути парною (тобто, непарна кількість точок).")

    h = (x[-1] - x[0]) / n
    integral = y[0] + y[-1]

    for i in range(1, n, 2):
        integral += 4 * y[i]

    for i in range(2, n-1, 2):
        integral += 2 * y[i]

    integral *= h / 3
    return integral

x = [-3.74, -3.60, -3.46, -3.32, -3.18, -3.04, -2.9, -2.76, -2.62, -2.48, -2.34, -2.2, -2.06, -1.92, -1.78, -1.64, -1.5, -1.36, -1.22, -1.08 ,-0.94 ,-0.8 ,-0.66 ,-0.52 ,-0.38 ,-0.24 ,-0.1]
y = [4.7, 5.16, 5.2, 5.5, 5.21, 5.32, 5.29, 5.18, 5.09, 5.18,  5.19, 4.97, 4.75, 4.48, 4.53, 4.4, 4.28, 4.5, 4.3, 4.05, 4, 3.9, 3.85, 3.83, 3.85, 4.03, 4.05]
manual_result = simpsons_rule(x, y)
scipy_result = simps(y, x)
print(f"Результат інтегрування методом Симпсона: {manual_result}")
print(f"Результат інтегрування за допомогою scipy: {scipy_result}")
if np.isclose(manual_result, scipy_result):
    print("Обидва результати співпадають.")
else:
    print("Результати не співпадають.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([3, 2.8, 2.6, 2.4, 2.2, 2, 1.8, 1.6, 1.4, 1.2, 1, 0.8, 0.6, 0.4, 0.2, 0])
y = np.array([1.6, 1.9, 2.1, 2.2, 2.64, 2.87, 2.77, 3.2, 3.7, 4.1, 4.2, 4.3, 4.1, 3.7, 4.27, 4.2])
def trapezoidal_integration(x, y):
    return np.sum((x[1:] - x[:-1]) * (y[:-1] + y[1:]) / 2)
integral_manual = trapezoidal_integration(x, y)
integral_numpy = np.trapz(y, x)
print(f"Інтеграл (вручну, метод трапецій): {integral_manual}")
print(f"Інтеграл (NumPy): {integral_numpy}")
```

#Метод прямокутників

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([2.8, 2.6, 2.4, 2.2, 2, 1.8, 1.6, 1.4, 1.2, 1, 0.8, 0.6, 0.4, 0.2, 0])
y = np.array([1.2, 1.2, 1.3, 1.5, 1.3, 1.16, 0.6, 0.4, 0.4, 0.3, 0.4, 0.55, 0.088, 0.01, 0])
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Графік по даним точкам')
plt.xlabel('Значення X')
plt.ylabel('Значення Y')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

```python
import numpy as np

def rectangle_method(x, y):
    dx = np.diff(x)
    area = np.sum(dx * np.abs(y[:-1]))
    return area

def simpson_method(x, y):
    if len(x) % 2 == 0:
        raise ValueError("Кількість підінтервалів повинна бути непарною")

    n = len(x) - 1
    h = (x[-1] - x[0]) / n
    area = np.abs(y[0]) + np.abs(y[-1])

    for i in range(1, n, 2):
        area += 4 * np.abs(y[i])
    for i in range(2, n - 1, 2):
        area += 2 * np.abs(y[i])

    area *= h / 3
    return area

x = np.array([2.8, 2.6, 2.4, 2.2, 2, 1.8, 1.6, 1.4, 1.2, 1, 0.8, 0.6, 0.4, 0.2, 0])
y = np.array([1.2, 1.2, 1.3, 1.5, 1.3, 1.16, 0.6, 0.4, 0.4, 0.3, 0.4, 0.55, 0.088, 0.01, 0])
area_rectangle = rectangle_method(x, y)
area_simpson = simpson_method(x, y)

print("Площа під кривою (метод середніх прямокутників):", area_rectangle)
print("Площа під кривою (метод Симпсона):", area_simpson)
```
