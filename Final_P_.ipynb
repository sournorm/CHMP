#Метод Сімпсона

```python
import matplotlib.pyplot as plt
x = [-2.4, -2.2, -2.0, -1.8, -1.6, -1.4, -1.2, -1.0, -0.8, -0.6, -0.4, -0.2, 0.0]
y = [2.13, 2.59, 2.62, 2.55, 2.51 , 4.79, 4.85, 4.9, 5.17, 5.21, 6.68, 7.09, 7.19]
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Графік по даним точкам')
plt.xlabel('Значення X')
plt.ylabel('Значення Y')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

```python
import numpy as np
from scipy.integrate import simps
def simpsons_rule(x, y):
    n = len(x) - 1
    if n % 2 == 1:
        raise ValueError("Кількість інтервалів повинна бути парною (тобто, непарна кількість точок).")

    h = (x[-1] - x[0]) / n
    integral = y[0] + y[-1]

    for i in range(1, n, 2):
        integral += 4 * y[i]

    for i in range(2, n-1, 2):
        integral += 2 * y[i]

    integral *= h / 3
    return integral

x = [-2.4, -2.2, -2.0, -1.8, -1.6, -1.4, -1.2, -1.0, -0.8, -0.6, -0.4, -0.2, 0.0]
y = [2.13, 2.59, 2.62, 2.55, 2.51 , 4.79, 4.85, 4.9, 5.17, 5.21, 6.68, 7.09, 7.19]
manual_result = simpsons_rule(x, y)
scipy_result = simps(y, x)
print(f"Результат інтегрування методом Симпсона: {manual_result}")
print(f"Результат інтегрування за допомогою scipy: {scipy_result}")
if np.isclose(manual_result, scipy_result):
    print("Обидва результати співпадають.")
else:
    print("Результати не співпадають.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1.80, 1.67, 1.54, 1.41, 1.28, 1.15, 1.02, 0.89, 0.76, 0.63, 0.50, 0.37, 0.24, 0.11])
y = np.array([3.70, 3.72, 3.89, 3.90, 6.40, 7.53, 7.47, 7.44, 7.56, 7.56, 7.50, 7.46, 7.42, 7.24])
def trapezoidal_integration(x, y):
    return np.sum((x[1:] - x[:-1]) * (y[:-1] + y[1:]) / 2)
integral_manual = trapezoidal_integration(x, y)
integral_numpy = np.trapz(y, x)

print(f"Інтеграл (вручну, метод трапецій): {integral_manual}")
print(f"Інтеграл (NumPy): {integral_numpy}")
```

```python
import numpy as np
x = np.array([-2.2, -2.0, -1.8, -1.6, -1.4, -1.2, -1.0, -0.8, -0.6, -0.4, -0.2, 0.0])
y = np.array([1.86, 1.92, 2.13, 1.33, 0.92, 0.93, 1.0, 0.98, 0.94, 0.95, 0.4, 0.0])
def trapezoidal_rule(x, y):
    integral = 0.0
    for i in range(1, len(x)):
        integral += (y[i] + y[i-1]) * (x[i] - x[i-1]) / 2
    return integral
integral_own = trapezoidal_rule(x, y)
integral_numpy = np.trapz(y, x)

print("Інтеграл (власна реалізація):", integral_own)
print("Інтеграл (numpy.trapz):", integral_numpy)
```

#Метод прямокутників

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1.93, 1.80, 1.67, 1.54, 1.41, 1.28, 1.15, 1.02, 0.89, 0.76, 0.63, 0.50, 0.37, 0.24, 0.11])
y = np.array([3.10, 2.71, 2.66, 2.35, 2.30, 2.16, 1.98, 1.43, 1.17, 1.10, 0.80, 0.17, 0.09, 0.04, 0.04])
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o')
plt.title('Графік по даним точкам')
plt.xlabel('Значення X')
plt.ylabel('Значення Y')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

```python
import numpy as np
def rectangle_method(x, y):
    dx = np.diff(x)
    area = np.sum(dx * y[:-1])
    return area
def simpson_method(x, y):
    if len(x) % 2 == 0:
        raise ValueError("Кількість підінтервалів повинна бути непарною.")

    n = len(x) - 1
    h = (x[-1] - x[0]) / n
    area = y[0] + y[-1]
    for i in range(1, n, 2):
        area += 4 * y[i]
    for i in range(2, n - 1, 2):
        area += 2 * y[i]

    area *= h / 3
    return area
x = np.array([1.93, 1.80, 1.67, 1.54, 1.41, 1.28, 1.15, 1.02, 0.89, 0.76, 0.63, 0.50, 0.37, 0.24, 0.11])
y = np.array([3.10, 2.71, 2.66, 2.35, 2.30, 2.16, 1.98, 1.43, 1.17, 1.10, 0.80, 0.17, 0.09, 0.04, 0.04])
area_rectangle = rectangle_method(x, y)
area_simpson = simpson_method(x, y)
print("Площа під кривою (метод середніх прямокутників):", area_rectangle)
print("Площа під кривою (метод Симпсона):", area_simpson)

```
